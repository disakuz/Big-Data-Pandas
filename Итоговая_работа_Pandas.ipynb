Задание 4
Добавьте к датафрейму из задания 3 столбцы:

```python
r_m = raiting.merge(movies,how = 'inner') # соединяем датафреймы
r_m.head()
```

```python
r_m['Total_count']=r_m.groupby('movieId')['movieId'].transform('count') # столбец с количеством оценок от всех пользователей
```

```python
r_m['Total_sum']=r_m.groupby('movieId')['movieId'].transform('sum') # # столбец с суммой оценок от всех пользователей
```

```python
r_m.head()
```

```python
r_m = r_m[['userId','movieId','rating','release_date', 'unknown', 'Action', 'Adventure',
 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western','Total_count', 'Total_sum']]
```
